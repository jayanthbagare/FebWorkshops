## Data Cleaning Operations

Your dataset can contain bad data that could be:

* Empty Values

* Wrong data

* Values in Wrong Format

* Duplicate data


For machine learning quality of data plays a vrey important role in getting desired outputs

Pandas offers several functions inorder to cleanse the data


### Empty Values

Empty values can potentially give you a wrong result when you analyze data.

One way is to <b>delete the rows containing empty values </b>, of this use the function df.dropna()

Syntax: df.dropna(inplace=True)

inplace=True, acts on the DataFrame and changes it in place


```python
df.dropna(inplace = True)
```


Another way of dealing with empty cells is to <b>insert a new value instead</b>

This way you do not have to delete entire rows just because of some empty cells

The fillna() method allows us to replace empty cells with a value

Syntax: df["Column_name"].fillna(value, inplace = True)



```python
df = pd.read_csv("Sales_Data.csv")
df.query('City.isnull()')
```

```python
df["City"].fillna("Bangalore",inplace = True)
df.query('City.str.contains("Bangalore")')
```

Another way of dealing is to <b>replace using Mean, Median, or Mode </b>

A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column


Syntax:

<b>Mean</b>

meanval = df["Column_Name"].mean()

df["Column_name"].fillna(meanval, inplace = True)

<br>
<b>Median</b>

medianval = df["Column_Name"].median()

df["Column_name"].fillna(medianval, inplace = True)
    
<br>
<b>Mode</b>

modeval = df["Column_Name"].mode()

df["Column_name"].fillna(modeval, inplace = True)    
    

```python
df = pd.read_csv("Sales_Data.csv")

meanval = df["SalesOrderID"].mean()

print(meanval)
```

```python
df["SalesOrderID"].fillna(meanval, inplace = True)
df.query('SalesOrderID =='+ str(meanval))
```

# Data Correlations


### Finding Relationships

A great aspect of the Pandas module is the corr() method

The corr() method calculates the relationship between each column in your data set

Use df.corr() to find the correlation

```python
df = pd.read_csv("Sales_Data.csv")
df.corr()
```

<b> Note: The corr( ) method ignores "not numeric" columns </b>

### Results 
   

The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.


### What is a good correlation? 

It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.
