```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("dark")
```

```python
# generate random numbers for each 0.5 step timepoint
df = pd.DataFrame(columns=["x","y"])
for i in range(100):
    x = np.random.randint(1,20)
    y = np.random.rand()
    df.loc[len(df)] = [x,y]
```

```python
xax = [*range(1,20)]
avg = []
for i in range(1,20):
    df1 = df.query(f"x == {i}")
    avg.append(df1['y'].mean())
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
p1 = sns.scatterplot(data=df, x='x', y='y', ax=ax)
p2 = sns.lineplot(x=xax, y=avg, color='r', ax=ax)
# p3 = sns.lmplot(data=df,x='x',y='y',line_kws={'color': 'y'}, height=3.5, aspect=1.5)
```
